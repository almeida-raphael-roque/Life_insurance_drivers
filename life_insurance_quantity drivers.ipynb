```python
import pandas as pd
import awswrangler as awr
import openpyxl
```

```python


path_to_query = r"C:\Users\raphael.almeida\Documents\Projetos\Seguro de Vida Motorista\listagem_mestra_ajustada.sql"

with open (path_to_query,'r') as reading:
    query_readed = reading.read()

df = awr.athena.read_sql_query(query_readed,'silver')
```

filtrando os conjuntos que tenham veículos pesados

```python


conjuntos_cascos = df[(df['beneficio'] == 'CASCO (VEÍCULO)')&
                      (df['tipo_veiculo'].str.contains('Pesado', na=False)|
                       df['tipo_veiculo'].str.contains('Micro-ônibus', na=False))
                    ]

lista_conjuntos_cascos = conjuntos_cascos['conjunto'].unique().tolist()

df_conjuntos_cascos = df[df['conjunto'].isin(lista_conjuntos_cascos)]
df_conjuntos_cascos['conjunto'].nunique()
```

filtrando apenas os conjuntos que tenham pendências de cadastramento

```python
# Filtrar conjuntos de df_conjuntos_cascos onde a quantidade de 'CASCO (VEÍCULO)' > quantidade de 'SEGURO DE VIDA' com 'MOTORISTA' na faixa_valor

# Contar 'CASCO (VEÍCULO)' por conjunto em df_conjuntos_cascos
casco_counts = df_conjuntos_cascos[df_conjuntos_cascos['beneficio'] == 'CASCO (VEÍCULO)'].groupby('conjunto').size()

# Contar 'SEGURO DE VIDA' com 'MOTORISTA' na faixa_valor por conjunto em df_conjuntos_cascos
vida_motorista_counts = df_conjuntos_cascos[
    (df_conjuntos_cascos['beneficio'] == 'SEGURO DE VIDA') &
    (df_conjuntos_cascos['faixa_valor'].str.contains('MOTORISTA', na=False))
].groupby('conjunto').size()

# Selecionar conjuntos onde a contagem de 'CASCO (VEÍCULO)' é maior que a de 'SEGURO DE VIDA' com 'MOTORISTA'
conjuntos_filtrados = [
    conjunto for conjunto in casco_counts.index
    if casco_counts.get(conjunto, 0) > vida_motorista_counts.get(conjunto, 0)
]

# Criar o dataframe filtrado
df_conjuntos_cascos_vida = df_conjuntos_cascos[df_conjuntos_cascos['conjunto'].isin(conjuntos_filtrados)]
df_conjuntos_cascos_vida['conjunto'].nunique()

```

conferências viavante

```python

df_cooperativa = df_conjuntos_cascos_vida[df_conjuntos_cascos_vida['cooperativa']=='Stcoop']
conjuntos_cooperativa = df_cooperativa['conjunto'].unique().tolist()
conjuntos_cooperativa
```
